```python
from bs4 import BeautifulSoup
import requests
```

```python
page = requests.get('https://www.imdb.com/list/ls091520106/')
```

```python
page
```

```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_title = soup.find(class_="lister-item-content")
first_title
```

```python
first_title.text
```

```python
first_title.text.replace('\n','')
```

```python
rating = []

for i in soup.find_all('div',class_="ipl-rating-widget"):
    rating.append(i.text.replace('\n',''))
    
rating
```

```python
location = []

for i in soup.find_all('div',class_="lister-item-year text-muted unbold"):
    location.append(i.text.replace('\n',''))
    
location
```

```python
titles = []

for i in soup.find_all('div',class_="lister-item-content"):
    titles.append(i.text)
    
titles
```

```python
titles = []

for i in soup.find_all('div',class_="lister-item-content"):
    titles.append(i.text.replace('\n',''))
    
titles
```

```python
import pandas as pd
df = pd.DataFrame({'titles':titles,'rating':rating})
df
```
