```python
soup = BeautifulSoup(page.content)
soup
```

```python
first_title = soup.find(class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full")
first_title
```

```python
first_title.text
```

```python
titles = []

for i in soup.find_all('div',class_="overflow-hidden overflow-ellipsis whitespace-nowrap max-w-80pe po:max-w-full"):
    titles.append(i.text)
    
titles
```

```python
area = []

for i in soup.find_all('div',class_="flex"):
    area.append(i.text)
    
area
```
